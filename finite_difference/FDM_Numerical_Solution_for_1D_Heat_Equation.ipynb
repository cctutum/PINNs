```python
import np as np
from matplotlib import plt as plt
```

```python
length = 10
k = 0.89
# B.C
temp_left = 100
temp_right = 200
total_sim_time = 10
```

```python
dx = .1
x_vector = np.linspace(0, length, int(length/dx))
```

```python
dt = 0.0001
t_vector = np.linspace(0, total_sim_time, int(total_sim_time/dt))
```

```python
# I.C
u = np.zeros([len(t_vector),len(x_vector)])
```

```python
u.shape
```

```python
u[0,0] =100
```

```python
u
```

```python
u[:,0] = temp_left
u[:,-1] = temp_right
```

```python
u
```

```python
plt.plot(x_vector,u[0])
plt.ylabel("temp")
plt.xlabel("x")
plt.show()
```

```python
for t in range(1, len(t_vector)-1):
    for x in range(1, len(x_vector)-1):
        u[t+1,x] = ((k* (dt/dx**2)) * (u[t, x+1] - 2* u[t,x] + u[t,x-1])) + u[t,x]
```

```python
u
```

```python
plt.plot(x_vector,u[0])
plt.plot(x_vector,u[100])
plt.plot(x_vector,u[1000])
plt.plot(x_vector,u[10000])
plt.plot(x_vector,u[100000-1])
plt.ylabel("temp")
plt.xlabel("x")
plt.show()
```
